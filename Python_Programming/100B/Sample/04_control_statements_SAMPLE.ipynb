# Python Control Statements — SAMPLE

Teaching/practice notebook for beginner Python.

## 1. Sequential Control Flow
By default, Python runs code from top to bottom.

```python
print("Step 1")
print("Step 2")
print("Step 3")
```

## 2. Conditional Statements

```python
x = 12
if x > 10:
    print("x is greater than 10")
```

## 3. If Statement and Indentation

```python
x = 5
if x > 0:
    print("x is positive")
    print("This line is inside the if block")
print("This line is outside the if block")
```

## 4. Comparison Operators

```python
a = 10
b = 5
print(a > b)
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```

## 5. Logical Operators

```python
age = 20
has_ticket = True
print(age >= 18 and has_ticket)
print(age < 18 or has_ticket)
print(not has_ticket)
```

## 6. If Else

```python
x = -3
if x > 0:
    print("x is positive")
else:
    print("x is not positive")
```

## 7. If Elif Else

```python
x = 0
if x > 0:
    print("x is positive")
elif x < 0:
    print("x is negative")
else:
    print("x is zero")
```

## 8. Nested If Statements

```python
x = 0
if x >= 0:
    if x > 0:
        print("x is positive")
    else:
        print("x is zero")
else:
    print("x is negative")
```

## 9. If Expressions

```python
x = 7
status = "positive" if x > 0 else "not positive"
print(status)
```

## 10. While Loop

```python
numbers = [1, 2, 3]
while numbers:
    print(numbers)
    numbers.pop(0)
```

## 11. For Loop

```python
for number in [1, 2, 3]:
    print(number)
```

## 12. Range Function

```python
for i in range(10):
    print(i)
```

```python
for i in range(2, 11, 2):
    print(i)
```

## 13. Break Statement

```python
for i in range(10):
    print(i)
    if i == 5:
        break
```

## 14. Continue Statement

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

## 15. Pass Statement

```python
for i in range(10):
    if i % 2 == 0:
        print(i)
    else:
        pass
```

## 16. Mini Project: Grade Classifier

```python
score = 76
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print(f"Score: {score}, Grade: {grade}")
```
