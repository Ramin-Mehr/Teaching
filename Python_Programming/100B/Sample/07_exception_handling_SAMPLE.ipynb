## 1. What is an Exception?
An exception is an error that happens while a program is running.

```
# Example: ZeroDivisionError
x = 10
y = 0

# Uncomment to see the error:
# print(x / y)
```

## 2. Handling Exceptions with try-except

```
try:
    result = 10 / 0
except ZeroDivisionError:
    print("You cannot divide by zero.")
```

## 3. Different Exception Types

```
value = "abc"

try:
    number = int(value)
    print(10 / number)
except ValueError:
    print("The value cannot be converted to an integer.")
except ZeroDivisionError:
    print("The number cannot be zero.")
```

## 4. Accessing the Exception Object

```
try:
    number = int("hello")
except ValueError as error:
    print("Error type:", type(error))
    print("Error message:", error)
```

## 5. else Clause

```
try:
    number = int("20")
except ValueError:
    print("Invalid number")
else:
    print("Conversion successful:", number)
```

## 6. finally Clause

```
try:
    file = open("temporary_demo.txt", "w")
    file.write("Hello")
except Exception as error:
    print("Something went wrong:", error)
finally:
    file.close()
    print("File closed.")
```

## 7. Raising Exceptions

```
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    return "Age is valid."

print(check_age(20))
```

## 8. Custom Exception

```
class InvalidAgeException(Exception):
    def __init__(self, age):
        self.age = age

    def __str__(self):
        return f"Invalid age: {self.age}. Age must be between 1 and 119."

def set_age(age):
    if age <= 0 or age >= 120:
        raise InvalidAgeException(age)
    print("Age accepted:", age)

try:
    set_age(150)
except InvalidAgeException as error:
    print(error)
```

## 9. Mini Example: Safe Calculator

```
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero."
    except TypeError:
        return "Error: Inputs must be numbers."

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, "x"))
```
