## 1. Pythonic Code
Python code should be readable and clear.

```
# Not very clear
x=10;y=20;print(x+y)

# Better
first_number = 10
second_number = 20
total = first_number + second_number
print(total)
```

## 2. Indentation: Use 4 Spaces

```
def greet(name):
    if name:
        print("Hello", name)

greet("Alice")
```

## 3. Naming Conventions

```
student_name = "John"      # variable: lowercase with underscores
MAX_SCORE = 100             # constant: uppercase

class StudentRecord:        # class: CapWords
    pass
```

## 4. Imports

```
# Standard library imports
import math
import random

print(math.sqrt(16))
print(random.randint(1, 5))
```

## 5. Whitespace

```
result = (3 + 4) * 2
numbers = [1, 2, 3, 4]

print(result)
print(numbers[1:3])
```

## 6. Comments

```
# Calculate the final price after discount.
price = 100
discount = 0.2
final_price = price * (1 - discount)

print(final_price)
```

## 7. Docstrings

```
def calculate_area(width, height):
    """Return the area of a rectangle."""
    return width * height

print(calculate_area(5, 4))
print(calculate_area.__doc__)
```

## 8. Type Hinting

```
def greeting(name: str) -> str:
    return "Hello " + name

print(greeting("Alice"))
```

## 9. Better String Concatenation

```
names = ["Alice", "John", "Sara"]
sentence = ", ".join(names)
print(sentence)
```

## 10. Refactoring Example

```
# Less readable
def f(x,y):
    return x*y

# More readable
def calculate_rectangle_area(width: float, height: float) -> float:
    """Return the area of a rectangle."""
    return width * height

print(calculate_rectangle_area(3, 4))
```
