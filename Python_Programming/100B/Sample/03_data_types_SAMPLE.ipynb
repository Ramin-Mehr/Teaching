# Python Data Types — SAMPLE

Teaching/practice notebook for beginner Python.

## 1. Type Function

```python
print(type(3))
print(type("Hello"))
print(type(3.5))
print(type(True))
```

## 2. Dynamic Typing

```python
x = 1
print(x, type(x))
x = "Hello"
print(x, type(x))
```

## 3. Numbers: int and float

```python
a = 3
b = -5
c = 3.5
print(type(a), type(b), type(c))
print(int("3"))
print(float("3.4"))
```

## 4. Strings

```python
s1 = "hi"
s2 = 'hi'
s3 = """This is
a multiline string."""
empty = ""
print(s1, s2)
print(s3)
print(len(empty))
```

## 5. String Operations

```python
s = "Hello World"
print(s + "!")
print(len(s))
print(s[0])
print(s[-1])
print(s[2:5])
print("Hi" * 3)
print(str(3))
```

## 6. String Methods

```python
text = " hello world "
print(text.upper())
print(text.strip())
print(text.replace("world", "Python"))
print(text.split())
print(text.count("l"))
print(text.find("world"))
name = "John"
print("The name is {}".format(name))
print(f"The name is {name}.")
```

## 7. Mutable vs Immutable

```python
my_list = [1, 2, 3]
my_list[0] = 4
print(my_list)
```

```python
my_string = "123"
# my_string[0] = "4"  # This creates an error because strings are immutable
new_string = "4" + my_string[1:]
print(new_string)
```

## 8. Boolean

```python
print(True)
print(False)
print(3 > 2)
print(bool(0))
print(bool(""))
print(bool("Python"))
print(int(True))
print(int(False))
```

## 9. None

```python
winner = None
print(winner is None)
winner = "Alice"
print(winner is not None)
```

## 10. Tuples

```python
t = (1, "Hello", True, (1, 3, 5))
print(t[0])
print(t[1:3])
print(len(t))
print(tuple([1, 2, 3]))
```

## 11. Lists

```python
names = ["John", "Paul", "George"]
names.append("Ringo")
print(names)
names.insert(1, "Alice")
print(names)
names.remove("Alice")
print(names)
print(names.pop(0))
print(names)
```

## 12. Sets

```python
s = set([1, 1, 2, 3])
print(s)
s.add(4)
s.update([5, 6, 7])
print(s)
print(1 in s)
```

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1 | s2)
print(s1 & s2)
print(s1 - s2)
print(s1 ^ s2)
```

## 13. Dictionaries

```python
capitals = {"Wales": "Cardiff", "England": "London"}
print(capitals["Wales"])
capitals["France"] = "Paris"
capitals["Wales"] = "Swansea"
print(capitals)
print("Wales" in capitals)
print(capitals.keys())
print(capitals.values())
print(capitals.items())
```

## 14. Choosing Collection Types

```python
# List: ordered items, duplicates allowed
students = ["Ali", "Sara", "Ali"]
# Set: unique items
unique_students = set(students)
# Dictionary: fast lookup by key
grades = {"Ali": 18, "Sara": 20}
print(students)
print(unique_students)
print(grades["Sara"])
```
