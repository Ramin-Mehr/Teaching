## 1. Working Directory and Paths

```
from pathlib import Path
print("Current directory:", Path.cwd())
```

## 2. Writing to a Text File

```
with open("students.txt", "w", encoding="utf-8") as file:
    file.write("Alice\n")
    file.write("John\n")
    file.write("Sara\n")

print("File created.")
```

## 3. Reading the Whole File

```
with open("students.txt", "r", encoding="utf-8") as file:
    content = file.read()

print(content)
```

## 4. Reading Line by Line

```
with open("students.txt", "r", encoding="utf-8") as file:
    for line in file:
        print(line.strip())
```

## 5. readlines()

```
with open("students.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

print(lines)
```

## 6. Appending to a File

```
with open("students.txt", "a", encoding="utf-8") as file:
    file.write("David\n")
```

## 7. File Attributes

```
file = open("students.txt", "r", encoding="utf-8")
print("Name:", file.name)
print("Mode:", file.mode)
print("Closed?", file.closed)
file.close()
print("Closed?", file.closed)
```

## 8. Rename and Delete Files

```
from pathlib import Path

old_file = Path("students.txt")
new_file = Path("class_students.txt")

if old_file.exists():
    old_file.rename(new_file)
    print("File renamed.")

print("Exists?", new_file.exists())

# Uncomment to delete:
# new_file.unlink()
```

## 9. Random Access with seek()

```
with open("letters.txt", "w+", encoding="utf-8") as file:
    file.write("abcdefghijklmnopqrstuvwxyz")
    file.seek(10)
    file.write("HELLO")
    file.seek(0)
    print(file.read())
```

## 10. Temporary Files

```
import tempfile

with tempfile.TemporaryFile(mode="w+", encoding="utf-8") as temp:
    temp.write("Temporary data")
    temp.seek(0)
    print(temp.read())
```
