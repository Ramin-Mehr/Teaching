## 1. Importing Built-in Modules

```
import math

print(math.sqrt(25))
print(math.pi)
```

## 5. Python Search Path

```
import sys

for path in sys.path[:5]:
    print(path)
```

## 6. Creating a Simple Module

```
module_code = '''
def greet(name):
    return f"Hello, {name}!"

def add(a, b):
    return a + b
'''

with open("my_utils.py", "w", encoding="utf-8") as file:
    file.write(module_code)

print("Module file created.")
```

```
import my_utils

print(my_utils.greet("Alice"))
print(my_utils.add(3, 4))
```

## 7. Importing Specific Items from Your Module

```
from my_utils import greet

print(greet("John"))
```

## 8. Script Pattern: if __name__ == '__main__'

```
script_code = '''
def main():
    print("This runs only when the file is executed directly.")

if __name__ == "__main__":
    main()
'''

with open("demo_script.py", "w", encoding="utf-8") as file:
    file.write(script_code)

print("Script file created.")
```
