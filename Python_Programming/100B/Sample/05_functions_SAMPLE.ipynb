# 05 - Functions: Sample Notebook

This notebook demonstrates Python functions for teaching: built-in functions, imports, user-defined functions, parameters, default values, `*args`, `**kwargs`, lambda functions, recursion, scope, and higher-order functions.

## 1. Built-in Functions
Built-in functions are available directly in Python.

```python
print("Hello, functions!")
print(abs(-12))
print(max(2, 5, 1))
print(len("Python"))
```

## 2. Importing Functions from Modules

```python
import math

print(math.exp(2))
print(math.sqrt(16))
print(dir(math)[:10])
```

```python
from math import pi, pow

print(pi)
print(pow(2, 3))
```

```python
import math as m

print(m.factorial(5))
```

## 3. Defining a Function

```python
def greet():
    """Print a simple greeting."""
    print("Welcome to Python!")

greet()
```

## 4. Function Parameters and Arguments

```python
def greet_student(name):
    print("Welcome", name)

greet_student("Alice")
greet_student("John")
```

## 5. Returning Values

```python
def add(a, b):
    return a + b

result = add(10, 5)
print(result)
```

## 6. Example: Factorial Function

```python
def factorial(n):
    """Return the factorial of a positive integer n."""
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
print(factorial.__doc__)
```

## 7. Default Parameter Values

```python
def introduce(name, course="Python Programming"):
    print(name, "is studying", course)

introduce("Alice")
introduce("John", "Data Science")
```

## 8. Named Arguments

```python
def student_info(name, age, major):
    print(f"Name: {name}, Age: {age}, Major: {major}")

student_info(age=21, major="Computer Science", name="Sara")
```

## 9. Arbitrary Positional Arguments: *args

```python
def greeter(*args):
    for name in args:
        print("Welcome", name)

greeter("John")
greeter("John", "Alice", "Sara")
```

## 10. Positional and Keyword Arguments: *args and **kwargs

```python
def show_data(*args, **kwargs):
    for arg in args:
        print("arg:", arg)
    for key, value in kwargs.items():
        print("key:", key, "value:", value)

show_data("John", "Denise", daughter="Phoebe", son="Adam")
```

## 11. Anonymous Functions: lambda

```python
double = lambda i: i * 2
square = lambda i: i * i

print(double(10))
print(square(5))
```

## 12. Recursion

```python
def recursive_factorial(n):
    if n == 1:
        return 1
    return n * recursive_factorial(n - 1)

print(recursive_factorial(5))
```

## 13. Local and Global Scope

```python
x = 100

def show_global():
    print("Inside function:", x)

show_global()
print("Outside function:", x)
```

```python
def local_example():
    local_value = 50
    print("Inside:", local_value)

local_example()
# Uncommenting the next line will cause an error because local_value is local
# print(local_value)
```

## 14. The global Keyword

```python
counter = 0

def increase_counter():
    global counter
    counter += 1

increase_counter()
increase_counter()
print(counter)
```

## 15. The nonlocal Keyword

```python
def outer():
    title = "original title"

    def inner():
        nonlocal title
        title = "another title"
        print("inner:", title)

    inner()
    print("outer:", title)

outer()
```

## 16. Function References and Higher-Order Functions

```python
def shout(text):
    return text.upper()

another_reference = shout
print(another_reference("hello"))
```

```python
def apply(x, my_function):
    return my_function(x)

def cube(n):
    return n ** 3

print(apply(3, cube))
print(apply(4, lambda x: x + 10))
```

## Mini Teaching Activity
Create a function that calculates the final grade based on assignment, midterm, and final exam scores.

```python
def final_grade(assignment, midterm, final_exam):
    return assignment * 0.3 + midterm * 0.3 + final_exam * 0.4

print(final_grade(80, 75, 90))
```
