# 06 - Classes: Sample Notebook

This notebook demonstrates object-oriented programming in Python: classes, objects, attributes, methods, `self`, constructors, `__str__`, class variables, class methods, static methods, encapsulation, and properties.

## 1. What is a Class?
A class is a template for creating objects.

```python
class Person:
    pass

p1 = Person()
print(type(p1))
```

## 2. The `__init__` Method and Attributes

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("John", 36)
p2 = Person("Phoebe", 21)

print(p1.name, p1.age)
print(p2.name, p2.age)
```

## 3. The `self` Parameter
`self` refers to the current object.

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def show_info(self):
        print(self.name, "has grade", self.grade)

s1 = Student("Alice", 90)
s1.show_info()
```

## 4. Assignment and Object References

```python
p1 = Person("John", 36)
px = p1

px.age = 40
print(p1.age)
print(px.age)
print(p1 is px)
```

## 5. Dot Notation

```python
print(p1.name)
p1.name = "Jonathan"
print(p1.name)
```

## 6. Printing Objects with `__str__`

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return self.name + " is " + str(self.age)

p1 = Person("John", 36)
print(p1)
```

## 7. Class Docstring

```python
class Person:
    """A simple class to represent a person with name and age."""

    def __init__(self, name, age):
        self.name = name
        self.age = age

print(Person.__doc__)
```

## 8. The `del` Keyword

```python
temp = Person("Temporary", 1)
print(temp.name)
del temp
# Uncommenting the next line will cause an error
# print(temp.name)
```

## 9. Intrinsic Attributes

```python
p1 = Person("John", 36)
print(Person.__name__)
print(Person.__dict__.keys())
print(p1.__class__)
print(p1.__dict__)
```

## 10. Class Variables

```python
class Person:
    population = 0

    def __init__(self, name, age):
        Person.population += 1
        self.name = name
        self.age = age

p1 = Person("John", 36)
p2 = Person("Alice", 22)
print(Person.population)
```

## 11. Class Methods

```python
class Person:
    population = 0

    @classmethod
    def increment_population(cls):
        cls.population += 1

    def __init__(self, name, age):
        Person.increment_population()
        self.name = name
        self.age = age

p1 = Person("John", 36)
p2 = Person("Alice", 22)
print(Person.population)
```

## 12. Static Methods

```python
class MathHelper:
    @staticmethod
    def add(a, b):
        return a + b

print(MathHelper.add(3, 4))
```

## 13. Private Attributes by Convention

```python
class Person:
    def __init__(self, age):
        self._age = age

p = Person(25)
print(p._age)  # Possible, but convention says: treat as private
```

## 14. Getter and Setter Methods

```python
class Person:
    def __init__(self, age):
        self.set_age(age)

    def get_age(self):
        return self._age

    def set_age(self, new_age):
        if isinstance(new_age, int) and new_age > 0 and new_age < 120:
            self._age = new_age
        else:
            print("Invalid age")

p = Person(30)
print(p.get_age())
p.set_age(40)
print(p.get_age())
p.set_age(-5)
```

## 15. Properties

```python
class Person:
    def __init__(self, age):
        self.age = age

    @property
    def age(self):
        """Return the person age."""
        return self._age

    @age.setter
    def age(self, new_age):
        if isinstance(new_age, int) and 0 < new_age < 120:
            self._age = new_age
        else:
            raise ValueError("Age must be an integer between 1 and 119")

p = Person(30)
print(p.age)
p.age = 35
print(p.age)
```

## Mini Teaching Activity
Create a `BankAccount` class with deposit and withdraw methods.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Not enough balance")

    def __str__(self):
        return f"{self.owner}: {self.balance} EUR"

account = BankAccount("Alice", 100)
account.deposit(50)
account.withdraw(30)
print(account)
```
