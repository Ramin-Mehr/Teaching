## 1. Expressions
An expression is code that produces a value.

```python
3 * 4
```

```python
3 * 4 + 10
```

```python
correct_area = 3 * 4
wrong_area = 3 ** 4
print(correct_area)
print(wrong_area)
```

## 3. Arithmetic Operators and Precedence

```python
print(3 + 4 * 2)
print((3 + 4) * 2)
print(10 / 3)
print(10 // 3)
print(10 % 3)
```

## 4. Assignment Operators

```python
score = 10
score += 5
print(score)
score *= 2
print(score)
```

## 5. Variables / Names

```python
a = 3 * 4
b = 2 * a
print(a)
print(b)
```

```python
user_name = "John"
print(user_name)
user_name = "Alice"
print(user_name)
```

## 6. Comparison, Logical, Identity, and Membership Operators

```python
x = 10
y = 5
print(x > y)
print(x == y)
print(x != y)
```

```python
age = 21
has_id = True
print(age >= 18 and has_id)
print(age < 18 or has_id)
print(not has_id)
```

```python
numbers = [1, 2, 3]
same_numbers = numbers
other_numbers = [1, 2, 3]
print(numbers == other_numbers)  # equality
print(numbers is other_numbers)  # identity
print(numbers is same_numbers)
print(2 in numbers)
```

## 7. Call Expressions and Built-in Functions

```python
print(abs(-12))
print(max(2, 2 + 3, 4))
print(min(8, 3, 10))
print(round(3.14159, 2))
```

```python
x = 2 + abs(-12)
print(x)
```
